### Comparing key attributes against house price

```python
plt.figure(figsize=(10,6), dpi=200)
sns.displot(train["SalePrice"])
```

```python
# SalePrice vs. Year Sold
sns.relplot(x="YrSold", y="SalePrice", data=train)
```

```python
# SalePrice vs. Overall Condition
sns.jointplot(x="OverallCond", y="SalePrice", data=train)
```

```python
# SalePrice vs. HeatingQC (Heating Quality and COndition)
sns.jointplot(x="HeatingQC", y="SalePrice", data=train)
```

```python
# SalePrice vs. Heating (Type of heating)
sns.jointplot(x="Heating", y="SalePrice", data=train)
```

```python
# SalePrice vs. Type of dwelling
sns.jointplot(x="BldgType", y="SalePrice", data=train)
```

```python
# SalePrice vs. MSSubClass (Type of dwelling)
sns.jointplot(x="MSSubClass", y="SalePrice", data=train)
```
